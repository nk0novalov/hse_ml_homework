```python
import csv
import json
import requests
IMDB_HOST = "https://imdb-api.kn-bc6.workers.dev/reviews/"
```

```python
csv_file_path = 'output_with_budget.csv'

# Read CSV file and convert to dictionary
data = []
with open(csv_file_path, newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    headers = next(csv_reader)  # Assumes the first row contains headers
    for row in csv_reader:
        entry = {}
        for i in range(len(headers)):
            entry[headers[i]] = row[i]
        data.append(entry)

# Serve data as a dictionary
json_data = json.dumps(data, indent=2)  # Convert the list of dictionaries to a JSON-formatted string
films_data = json.loads(json_data)  # Convert the JSON string back to a Python dictionary
```

```python
def get_comment_by_film_imdb_id(imdb_id: str):
    return requests.get(f"{IMDB_HOST}{imdb_id}?option=rating&sortOrder=asc").json()

def extract_reviews_by_imdb_id(imdb_id: str):
    try:
        data_input = get_comment_by_film_imdb_id(imdb_id)
        return data_input.get("reviews")
    except:
        return None
    

result = []

i = 0
cnt = 0
for film in films_data:
    cnt += 1
    if cnt >= 5912:
        print(cnt)
        film_id = film["id"]
        data_review = extract_reviews_by_imdb_id(film_id)
        if None:
            break
        if film_id:
            for review in data_review:
                result.append({"index": i, "movie": film['title'], "user_review_permalink": f"{review['reviewLink']}", "user_review": review['content'], "movie_id": f"/title/{film_id}/"})
                i += 1
```

```python
csv_file_path = 'review3.csv'

# Write data to CSV file
with open(csv_file_path, 'w', newline='') as csvfile:
    fieldnames = result[0].keys()  # Assuming all dictionaries have the same keys
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()  # Write the header row

    for row in result:
        writer.writerow(row)

print(f'Data written to CSV file: {csv_file_path}')
```
